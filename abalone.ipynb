## Task 3
Generate a single plot showing the (empirical) cumulative distribution function for the number of rings for each of the three sexes (male, female, infant).

```python

df=df_abalone[['Sex','Rings']]
df
```

```python
df_male = df[df['Sex'].str.contains('M')]
df_male
```

```python
df_female = df[df['Sex'].str.contains('F')]
df_female
```

```python
df_infant = df[df['Sex'].str.contains('I')]
df_infant
```

```python
from itertools import accumulate

fig,ax = plt.subplots()
for df in [df_male,df_female,df_infant]: 
    xh, xb = np.histogram(df['Rings'], bins=60, normed=True)
    xh = list(accumulate(xh))
    xh = np.array(xh) / max(xh)

    ax.plot(xb[1:], xh, label=f"$CDF$({df.iloc[0,0]})")
ax.legend()
plt.title("CDFs of Sexes")
```
